1. Attendance Data

    Read the data from the `attendance` table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.
    
    You should end up with something like this:

        name
        Billy    0.5250
        Jane     0.6875
        John     0.9125
        Sally    0.7625
        Name: grade, dtype: float64


```python
untidy_df = untidy_df.rename(columns={'Unnamed: 0': 'name'})

untidy_df
```

```python
untidy_df = untidy_df.melt(id_vars=['name'], var_name='date', value_name='attendance')
```

```python
untidy_df.head()
```

```python
untidy_df.attendance = untidy_df.attendance.map({'P':1, 'H':.5, 'A':0, 'T':.9})

untidy_df
```
